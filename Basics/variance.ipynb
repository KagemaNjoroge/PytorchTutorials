# Variance
Population variance is a measure of how spread out a group of data points is. Specifically, it quantifies the average squared deviation from the mean. So, if all data points are very close to the mean, the variance will be small; if data points are spread out over a wide range, the variance will be larger.  
`σ² = Σ (xi - μ)² / N`

```python
data = [32, 111, 138, 28, 59, 77, 97]
```

```python
#implementing variance function from scratch

def calculate_variance(data: list[int])->float:
    number_of_data_points = len(data)
    mean = sum(data) / number_of_data_points

    sum_of_distances_from_mean_squared = 0

    for point in data:
        distance = (point - mean)**2

        sum_of_distances_from_mean_squared += distance
    
    variance = sum_of_distances_from_mean_squared / number_of_data_points

    return variance

calculate_variance(data)

```

```python
# using numpy module

import numpy as np
variance = np.var(data)

variance
```
