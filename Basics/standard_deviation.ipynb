# Standard Deviation  
Standard deviation is a number that describes how spread out the values are.

A low standard deviation means that most of the numbers are close to the mean (average) value.

A high standard deviation means that the values are spread out over a wider range  

`σ = √(σ²)`

`standard_deviation = √(variance)`

```python
data = [6,2,3,1]
```

```python
# implementing stardard deviation function from scratch
def calculate_standard_deviation(data: list[int])-> float:
    number_of_data_points = len(data)
    mean = sum(data) /number_of_data_points
    sum_of_square_of_distances_from_mean = 0

    for point in data:
        distance = (point - mean)**2
        sum_of_square_of_distances_from_mean += distance
    
    standard_dev = ((sum_of_square_of_distances_from_mean) / number_of_data_points)**0.5

    return standard_dev
calculate_standard_deviation(data)
    
```

```python
# using numpy
import numpy as np
standard_deviation = np.std(data)

standard_deviation
```
